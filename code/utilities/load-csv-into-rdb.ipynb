## Description
Load a csv into a pandas dataframe and persist to a mysql db  

```python
import pandas as pd
import csv 
from pandas.io import sql
from sqlalchemy import create_engine
from sqlalchemy import types
```

```python
PATENT_DATA = '/Users/sanjay/dev/EAGER/data/patents/'
ENGINE = create_engine('mysql://root:innovate@localhost/patent_20170808?charset=utf8', encoding='utf-8')
CON = ENGINE.connect()
```

```python
table_name = 'eager_assignee'
file = PATENT_DATA + table_name + '.csv'
print (file)
table_df = pd.read_csv(file, sep=',', encoding='utf-8', warn_bad_lines=True, error_bad_lines=False)
table_df.head()
```

```python
max(table_df['firm'].apply(len))
```

```python
table_df.to_sql(name=table_name,con=CON,if_exists='replace', index=False, 
                dtype={'lookup_firm': types.NVARCHAR(length=256)}
               )
```

```python
# Setting up database connection information
CONNECTION_STRING = 'mongodb://localhost'
MONGODB_DB = "FirmDB_20181116"
TARGET_COLLECTION = "pages_COMBINED"
OUTF = "../../data/analysis/measures/simple_web_measures_v2.1.csv"

client = pymongo.MongoClient(CONNECTION_STRING)
db = client[MONGODB_DB]

pp = pprint.PrettyPrinter()
```

```python
words_by_firm_name = defaultdict(int)
pages_by_firm_name = defaultdict(int)

def get_measures():
    target_col = db[TARGET_COLLECTION]
    results = target_col.find ()
    
    for result in results:
        # print ('\tWorking on', result['url'][0])
        # pp.pprint (result)
        if 'firm_name' not in result:
            continuecombined)
        firm_name = result['firm_name'][0]
        pages_by_firm_name[firm_name] += 1
        if 'full_text' in result:
             words_by_firm_name[firm_name] += sum([len(text) for text in result['full_text']])
        # else continue
        
# standard firm cleaning regex
def clean_firm_name (firm):
    firm_clnd = re.sub('(\.|,| corporation| incorporated| llc| inc| international| gmbh| ltd)', '', firm, flags=re.IGNORECASE).rstrip()
    return firm_clnd
```

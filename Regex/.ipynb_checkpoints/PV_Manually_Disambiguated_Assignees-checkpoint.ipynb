```python
import sys
import pandas as pd
import numpy as np
```

```python
top_orgs = pd.read_csv("top_orgs_to_check_checked.csv")
top_orgs.head()
```

```python
assignees_lookedup_types = pd.read_csv("assignees_lookedup_types.csv")
assignees_lookedup_types.head()
```

```python
dfmerged = pd.merge(top_orgs, assignees_lookedup_types, on = "organization", how = "outer")
```

```python
dfmerged.head()
```

```python
dfmerged['final_type'] = dfmerged.apply(lambda row : row['updated_type'] if not pd.isnull(row['updated_type']) else row['thes_types'],axis=1) 
```

```python
dfmerged[~pd.isnull(dfmerged['updated_type'])].head(30)
```

```python
dfmerged.to_csv('merged_org_types.csv')
```
